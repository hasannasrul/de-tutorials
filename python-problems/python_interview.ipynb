### sort without sort function

```python
def quickSort(arr):
    if len(arr) <=1:
        return arr
    else:
        pivot = arr[0]
        leftarr = [i for i in arr[1:] if i < pivot]
        rightarr = [j for j in arr[1:] if j > pivot]

        return quickSort(leftarr) + [pivot] + quickSort(rightarr)

arr = [23,45,6,1,2,19,22]
print(quickSort(arr))
```

### Two Sum

```python
def two_sum(arr, k):
    seen = {}
    for i, num in enumerate(arr):
        diff = k - num
        if diff in seen:
            return [seen[diff], i]
        seen[num] = i
    return []

k = 9
arr = [2,3,5,7]
print(two_sum(arr,k))
    
```

Longest Substring

```python

def longest_substring(s):
    seen = {}
    left = 0
    max_len = 0

    for i, char in enumerate(s):
        if char in seen and seen[char] >= left:
            left = seen[char] +1
        
        seen[char] = i
        max_len = max(max_len, i-left+1)
    
    return max_len

print(longest_substring("dscsdsdavfs"))
```

fibonacci memoization

```python
def fibonacci(n, dct={}):
    if n in dct:
        return dct[n]
    if n<=1:
        return n

    dct[n] = fibonacci(n-1, dct) + fibonacci(n-2, dct)
    return dct[n]

for i in range(12):
    print(fibonacci(i), end=" ")  

```

Factorial using memoization

```python
def factorial(n, dct={}):
    if n in dct:
        return dct[n]
    if n <=1:
        return 1
    
    dct[n] = n * factorial(n-1, dct)
    return dct[n]


print(factorial(5))
    
```

### Flatten the dict

```python
unflat_json = {'user':
               {'Rachel':
                {'UserID': 1717171717,
                 'Email': 'rachel1999@gmail.com',
                 'friends': ['John', 'Jeremy', 'Emily']
                 }
                }
               }
```

```python
def flatten_json(x):
    result = {}

    def _flatten(y, name=''):
        if isinstance(y, dict):
            for k, v in y.items():
                new_key = f"{name}_{k}" if name else k
                _flatten(v, new_key)
        elif isinstance(y, list):
            for i,v in enumerate(y):
                new_key = f"{name}_{i}"
                _flatten(v, new_key)
        else:
            result[name] = y
    
    _flatten(x)
    return result

print(flatten_json(unflat_json))
```

### flatten list

```python
ls = [2,3,[1, 3, 5], [2, 4, 6]]


result = []
def flatten_list(ls):
    
    for i in ls:
        if isinstance(i, list):
            flatten_list(i)
        else:
            result.append(i)
    return result

print(flatten_list(ls))
```

### Parse List of strings to json

```python
logs = [
    "2025-07-18 12:00:01 GET /index.html 200",
    "2025-07-18 12:00:02 POST /login 401",
    "2025-07-18 12:00:03 GET /dashboard 200",
    "2025-07-18 12:00:04 GET /index.html 200",
    "2025-07-18 12:00:05 POST /login 200",
    "2025-07-18 12:00:05 PUT /login 200"
]

'''
{'GET': {'200': 3},'POST': {'401': 1, '200': 1}}
'''
```

```python
result = {}
for i in logs:
    log = i.split(" ")
    response = log[2]
    status = log[-1]

    if response not in result:
        result[response] = {}
    if status not in result[response]:
        result[response][status] = 0
    
    result[response][status]+=1

print(result)
```

### three sum problem
A + B + C = 0

find index of elements whose sum is equal to zero

```python
arr = [2,10,1,9,-1,0]

def three_sum(arr):
    total_len = len(arr)
    for i in range(0, total_len-2):
        for j in range(i+1,total_len-1):
            for k in range(j+1, total_len):
                if arr[i]+arr[j]+arr[k]==0:
                    return [i,j,k]


print(three_sum(arr))
```

```python
def three_sum_opt(a, target):
    n = len(a)
    for i in range(n):
        seen = {}
        current_target = target - a[i]   # what the remaining two numbers must sum to

        for j in range(i + 1, n):
            diff = current_target - a[j]

            if diff in seen:
                # Found the pair that, along with a[i], gives the target
                return [i, seen[diff], j]

            seen[a[j]] = j

    return None

arr = [2,10,1,9,-1,0]
result = three_sum_opt(arr, 0)
print(result)
```
