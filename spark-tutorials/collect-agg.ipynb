```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("count-null").getOrCreate()

data = [
    ("john", "tomato", 2),
    ("bill", "apple", 2),
    ("john", "banana", 2),
    ("john", "tomato", 3),
    ("bill", "taco", 2),
    ("bill", "apple", 2),
]
schema = ["name", "item", "weight"]
df = spark.createDataFrame(data, schema)
```

```python
df.show()
```

```python
df = df.groupBy("name", "item").agg(sum("weight").alias("total_weight"))
```

```python
df.show()
```

```python
result = df.groupBy("name").agg(collect_set(struct("item", "total_weight")).alias("item_weight"))
result.show()
```
