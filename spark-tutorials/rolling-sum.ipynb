### Find rolling three month average per customer
Sample Data

sale_id	sale_date	customer_id	amount
1	    2025-01-15	101	        100
2	    2025-02-10	101	        150
3	    2025-03-05	101	        200
4	    2025-03-20	102	        300
5	    2025-04-01	101	        250
6	    2025-04-15	102	        100
7	    2025-05-10	101	        300
8	    2025-05-20	102	        200


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window

spark = SparkSession.builder.appName("rolling-sum").getOrCreate()
```

```python
data = [
    (1,"2025-01-15", 101, 100),
    (2,"2025-02-10", 101, 150),
    (3,"2025-03-05", 101, 200),
    (4,"2025-03-20", 102, 300),
    (5,"2025-04-01", 101, 250),
    (6,"2025-04-15", 102, 100),
    (7,"2025-05-10", 101, 300),
    (8,"2025-05-20", 102, 200),
]
cols = ["sale_id", "sale_date", "customer_id", "amount"]
df = spark.createDataFrame(data, cols)
df.show()
```

```python
wspec = Window.partitionBy("customer_id").orderBy("sale_date")

df_rolling_only = df.withColumn("rolling_sum", sum(col('amount')).over(wspec))

df_3month = df.withColumn("rolling_3_month_avg", avg(col('amount')).over(wspec.rowsBetween(Window.currentRow-2,Window.currentRow)))
```

```python
df_3month.show()
df_rolling_only.show()
```
