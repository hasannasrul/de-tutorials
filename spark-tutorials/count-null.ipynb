```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("count-null").getOrCreate()

data = [
    ("Alice", None, 30),
    ("Bob", 25, None),
    (None, 35, 40),
    ("Charlie", 40, 50)
]
columns = ["Name", "Age", "Score"]
df = spark.createDataFrame(data, columns)
```

```python
df.show()
```

```python
null_df = df.select([ count(when(col(i).isNull(), i)).alias(i) for i in df.columns])
null_df.show()
```
